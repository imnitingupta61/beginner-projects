# Scaling the data using min max scaler

```python
from sklearn.preprocessing import MinMaxScaler

mn=MinMaxScaler()
x=mn.fit_transform(x)
```

# Importing Models

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

```python
lr= LinearRegression()
ne = KNeighborsRegressor(n_neighbors=2)
dt = DecisionTreeRegressor()
rf= RandomForestRegressor(n_estimators=100)
ab= AdaBoostRegressor()
rd= Ridge()
mod = [lr,ne, dt, rf, ab, rd]
mod
```

```python
for j in mod:
    for i in range(0,100):
        xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.22, random_state=i)
        j.fit(xtrain,ytrain)
        pred_train = j.predict(xtrain)
        pred_test= j.predict(xtest)

        print(f"For Model {j} At random state {i}, the training accuracy is: {r2_score(ytrain,pred_train)}")
        print(f"For Model {j} At random state {i}, the testing accuracy is: {r2_score(ytest,pred_test)}")
        print('\n')
```

As per above visualization we found model Random Forest Regressor is best at random state is 23

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.22, random_state=23)
predrf_train = rf.predict(xtrain)
predrf_test= rf.predict(xtest)

print("Mean Squared error", mean_squared_error(ytest,predrf_test))
print("Mean Absolute Error", mean_absolute_error(ytest,predrf_test))
```

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor()
param_grid = {"n_estimators":[50,100,200,500]}
xg_grid = GridSearchCV(estimator = model, param_grid = param_grid)
print(xg_grid)
xg_grid.fit(xtrain, ytrain)
```

```python
print(xg_grid.best_score_)
print(xg_grid.best_estimator_)
print(xg_grid.best_params_)
```

As per Grid Search CV we found the best accuracy when n_estimators=500, random_state= 23

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.22, random_state=23)
RFR = RandomForestRegressor(n_estimators=500)
RFR.fit(xtrain,ytrain)
predrfr_train =RFR.predict(xtrain)
predrfr_test= RFR.predict(xtest)
print("Mean Squared error", mean_squared_error(ytest,predrfr_test))
print("Mean Absolute Error", mean_absolute_error(ytest,predrfr_test))
```

```python
RFR.score(xtrain,ytrain)
```

# Cross Valdation of Model

```python
train_accuracy = r2_score(ytrain,predrfr_train)
test_accuracy =r2_score(ytest,predrfr_test)

from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cv_score =cross_val_score(RFR,x,y,cv=j)
    cv_mean = cv_score.mean()
    print(f"At cross val  {j} the cv score  is {cv_mean} and accuracy score for training is {train_accuracy} and accuracy for testing is {test_accuracy}")
    print("\n")
```
